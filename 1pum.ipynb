```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = '/content/Salary Data (2).csv'
try:
    salary_df = pd.read_csv(file_path)
    print("Nagłówek danych:")
    display(salary_df.head())

    cols = salary_df.columns.tolist()

    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=salary_df, x=cols[0], y=cols[-1])
    plt.title(f'Wykres punktowy: {cols[0]} vs {cols[-1]}')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

except Exception as e:
    print(f"Błąd podczas wczytywania lub rysowania: {e}")
```
